92페이지 연습문제 1

다음 코드를 사용하여 1일부터 7일까지의 뉴욕 타임즈 웹페이지 방문 로그 데이터를 다운로드하자.

```python
import matplotlib.pyplot as plt
import pandas as pd
day1 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt1.csv")
day2 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt2.csv")
day3 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt3.csv")
day4 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt4.csv")
day5 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt5.csv")
day6 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt6.csv")
day7 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt7.csv")
```

Age: 방문자의 나이 Gender: 방문자의 성별 (0, 1) Impressions: 노출된 기사의 수 Clicks: 클릭한
기사의 수 Signed_In: 로그인 한 경우에는 1, 그렇지 않은 경우는 0 (로그인을 하지 않은 방문자의 Age,
Gender는 0)

(1) 1일부터 7일까지의 일별 전체 방문자 수를 나타내는 그래프를 그리시오.

```python
x = [1,2,3,4,5,6,7]
y = [len(day1),len(day2),len(day3),len(day4),len(day5),len(day6),len(day7)]
l = plt.plot(x, y, 'ko', label='Visitors',linestyle='-')
plt.setp(l, markersize=10)
plt.setp(l, markerfacecolor='00')
plt.xlabel("Dates")
plt.ylabel("Visitors")
plt.title("Number of visitors per day by day")
plt.legend(loc='best')
plt.show()
```

(2) 1일부터 7일까지의 로그인 한 사람과 로그인 하지 않은 사람의 일별 평균 Click Through Rate (CTR) 을 구하고 이를 그래프로 나타내시오.

```python
day1['CTR'] = day1.Clicks / day1.Impressions
day2['CTR'] = day2.Clicks / day2.Impressions
day3['CTR'] = day3.Clicks / day3.Impressions
day4['CTR'] = day4.Clicks / day4.Impressions
day5['CTR'] = day5.Clicks / day5.Impressions
day6['CTR'] = day6.Clicks / day6.Impressions
day7['CTR'] = day7.Clicks / day7.Impressions
avgCTR = [day1.CTR.mean(),day2.CTR.mean(),day3.CTR.mean(),day4.CTR.mean(),day5.CTR.mean(),day6.CTR.mean(),day7.CTR.mean()]
l2 = plt.plot(x, avgCTR, 'ko', label='CTRavg',linestyle='-')
plt.setp(l2, markersize=10)
plt.setp(l2, markerfacecolor='00')
plt.xlabel("Dates")
plt.ylabel("Average of CTR")
plt.title("Average of CTR per day by day")
plt.legend(loc='best')
plt.show()
```

97페이지 연습문제 2

다음 코드를 사용하여 하루 동안의 뉴욕 타임즈 웹페이지 방문 로그 데이터를 다운로드하자.

```python
day1 = pd.read_csv("http://stat.columbia.edu/~rachel/datasets/nyt1.csv")
```

Age: 방문자의 나이 Gender: 방문자의 성별 (0, 1) Impressions: 노출된 기사의 수 Clicks: 클릭한
기사의 수 Signed_In: 로그인 한 경우에는 1, 그렇지 않은 경우는 0 (로그인을 하지 않은 방문자의 Age,
Gender는 0)

(1) 로그인한 사용자와 그렇지 않은 사용자 수를 비교하는 그래프를 그리시오.

```python
day1.Signed_In = pd.Categorical(day1.Signed_In)
t = day1.Signed_In.value_counts()
plt.bar(t.index,t.values,color=['blue', 'red'])
plt.show()
```

(2) 로그인한 사용자 중 성별이 0인 사용자와 1인 사용자수를 비교하는 그래프를 그리시오.

```python
day1.Gender = pd.Categorical(day1.Gender)
x = day1[day1.Signed_In == 1]
t = day1.Gender.value_counts()
plt.bar(t.index,t.values,color=['blue', 'red'])
plt.show()
```
